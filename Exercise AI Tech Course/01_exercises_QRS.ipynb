### Exercise 02: Arithmetic Operations Exercise
#### Task: Calculate and print the sum, difference, product, and quotient of two numbers provided by the user.

```python
# Your solution:
num1 = float(input("Please enter the first number: "))
num2 = float(input("Please enter the second number: "))

summ = num1 + num2
subtr = 0
if num1 >= num2:
    subtr = num1 - num2
else:
    subtr = num2 - num1
multp = num1 * num2 
quotn = num1 / num2

print(f"The Arithmatic results from the inputed numbers for summation is {summ}, the difference is {round(subtr, 1)}, the product of them is {multp} and quotient is {round(quotn, 3)}.")
```

### Exercise 06: Function Definition Exercise
#### Task: Define a function that takes two numbers and returns their average.

```python
# Your solution:
def avg(num1, num2):
    res = (num1+num2)/2
    return res

val1 = float(input("Please enter the first number: "))
val2 = float(input("Please enter the second number: "))

print("The average of those numbers is:", avg(val1, val2))
```
